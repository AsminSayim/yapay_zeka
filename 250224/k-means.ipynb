```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# İris veri setini yükle
iris = load_iris()
X = iris.data  # Özellikler (sepal ve petal uzunluğu, genişliği)
y = iris.target  # Etiketler (3 tür çiçek)

# 2. Adım: Veriyi ölçeklendir (özellikler farklı aralıklarda olabiliyor)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Adım: K-means algoritmasını uygula (k=3, çünkü İris veri seti 3 tür içeriyor)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# 4. Adım: Sonuçları görselleştir
# Kümeleri ve merkezleri görselleştireceğiz
plt.figure(figsize=(8, 6))

# Kümeleri renkli şekilde çiz
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k')

# Kümelerin merkezlerini kırmızı "X" ile göster
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200, label='Centroids')

plt.title('K-means Clustering on Iris Dataset')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.legend()
plt.show()
```
