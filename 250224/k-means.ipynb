```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ris veri setini ykle
iris = load_iris()
X = iris.data  # zellikler (sepal ve petal uzunluu, genilii)
y = iris.target  # Etiketler (3 tr iek)

# 2. Adm: Veriyi leklendir (zellikler farkl aralklarda olabiliyor)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Adm: K-means algoritmasn uygula (k=3, nk ris veri seti 3 tr ieriyor)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# 4. Adm: Sonular grselletir
# Kmeleri ve merkezleri grselletireceiz
plt.figure(figsize=(8, 6))

# Kmeleri renkli ekilde iz
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k')

# Kmelerin merkezlerini krmz "X" ile gster
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200, label='Centroids')

plt.title('K-means Clustering on Iris Dataset')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.legend()
plt.show()
```
