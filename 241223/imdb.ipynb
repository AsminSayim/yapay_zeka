```python
# Gerekli kütüphaneleri yükleyin
import pandas as pd
```

```python
# 1. Veri setini yükleyin
df = pd.read_csv('imdb_top_1000.csv')
```

```python
#Veri setinin bilgilerini yazdırın
df.info()
```

```python
#Null verileri kontrol edin ve hangi sütunlarda null veri olduğunu gösterin
df.isnull().sum()
```

```python
# Boş (null) verileri veri setinden kaldırın
df.dropna(subset=['Certificate','Meta_score','Gross'], inplace=True)
```

```python
#Belirtilen sütunları veri setinden kaldırın
df = df.drop(columns=["Poster_Link", "Certificate", "Overview", "Star1", "Star2", "Star3", "Star4"])
```

```python
#Sütun başlıklarını Türkçeye çevirin
df = df.rename(columns={
    "Series Title": "Film Adı",
    "Genre": "Tür",
    "IMDB Rating": "IMDB Puanı",
    "Votes": "Oy Sayısı",
    "Gross": "Hasılat",
    "Director": "Yönetmen",
    "Stars": "Yıldızlar",
    "Runtime": "Süre"
})
```

```python
#Türü 'Drama' olan satırları gösterin
drama_df = df[df['Tür'] == 'Drama']
drama_df
```

```python
#IMDB Puanı 8.5'ten fazla olan satırları gösterin
high_rating_df = df[df['IMDB_Rating'] > 8.5]
high_rating_df
```
